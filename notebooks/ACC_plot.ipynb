```python
# rende config.py (in notebooks/) importabile anche dai notebook in sottocartelle
import sys, os
_cfg = os.getcwd()
while _cfg != os.path.dirname(_cfg):
    if os.path.exists(os.path.join(_cfg, 'config.py')):
        sys.path.insert(0, _cfg)
        break
    _cfg = os.path.dirname(_cfg)
from config import CONFESS_DATA, BC_DATA, ERA5_ROOT, POST_DATA, WORK_DIR, FIG_DIR, FIG_DIR_2025
import os
import albedo_functions as af
import xarray as xr
import matplotlib.pyplot as plt
from matplotlib.colors import LinearSegmentedColormap
import logging
from pathlib import Path
```

```python
# Definizione di una colormap asimmetrica da blu a bianco a rosso
colors = [
    (0, 0.3, 0.75),  # Blu meno intenso
    (1, 1, 1),      # Bianco al centro
    (0.8, 0, 0)     # Rosso più intenso
]

# Definire la posizione relativa (asimmetrica) di ciascun colore
positions = [0.0, 0.5, 1.0]  # Bianco centrato a metà

# Creare la colormap personalizzata
custom_cmap = LinearSegmentedColormap.from_list("custom_cmap", list(zip(positions, colors)))
```

```python
# Configura il logging
log_file = "ACC_plot.log"
logging.basicConfig(
    level=logging.INFO,
    format="%(asctime)s - %(levelname)s - %(message)s",
    handlers=[
        logging.FileHandler(log_file),  # Write to file
#        logging.StreamHandler()  # Still see some output in Jupyter
    ]
)

# Percorso ai dati
DATA_PATH = POST_DATA
OUTPUT_PATH = FIG_DIR_2025
OBS_PATH = WORK_DIR

# Esperimenti e variabili
exp_sens = "a52o"
exp_ctrl = "a1ua"
var = "tas"

season = 'DJF'
y1=3
y2=4
lead = f"{y1}-{y2}"
lead_number = y2 - y1 + 1
```

```python
ctrl_acc = xr.open_dataset(f"{OBS_PATH}/{exp_ctrl}_{var}_ACC_r_{lead}_{season}_1999_oce.nc")
#ctrl_acc = af.land_seas_mask(ctrl_acc)
ctrl_p = xr.open_dataset(f"{OBS_PATH}/{exp_ctrl}_{var}_ACC_p_{lead}_{season}_1999_oce.nc")
#ctrl_p = af.land_seas_mask(ctrl_p)
```

```python
sens_acc = xr.open_dataset(f"{OBS_PATH}/{exp_sens}_{var}_ACC_r_{lead}_{season}_1999_oce.nc")
#sens_acc = af.land_seas_mask(sens_acc)
sens_p = xr.open_dataset(f"{OBS_PATH}/{exp_sens}_{var}_ACC_p_{lead}_{season}_1999_oce.nc")
#sens_p = af.land_seas_mask(sens_p)
```

```python
delta = sens_acc - ctrl_acc
sig_delta = xr.open_dataset(f"{OBS_PATH}/{exp_sens}-{exp_ctrl}_{var}_bootstrap_{lead}_{season}_1999_oce.nc")
#sig_delta = af.land_seas_mask(sig_delta)
```

```python
title = f"{exp_ctrl}_{var}_acc_lead_{lead}_{season}_1999_colorbar_oce"
output_file = OUTPUT_PATH / f"{title}.png"
af.map_plot(ctrl_acc[f"{var}"], ctrl_p[f"{var}"], levels, title = title,cmap = custom_cmap, sign = 0.95, antartica=False)
plt.savefig(output_file, dpi = 600, bbox_inches = 'tight')
```
